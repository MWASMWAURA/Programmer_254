```python
#Bank manager
import sys
class BankAccount:
    def __init__(self,account_holder,balance=0.00):
        self.account_holder=account_holder
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f"You have deposited ${amount}, your account balance is ${self.balance}")
    def withdraw(self,amount):
        self.balance-=amount
        print(f"You have withdrawn ${amount}, your account balance is ${self.balance}")
    def get_balance(self):
        return self.balance
    
class Bank:
    def __init__(self,name):
        self.name=name
        self.accounts=[]
    def add_account(self,account):
        self.accounts.append(account)
        print(f"Account for {account.account_holder} added to the bank.")

    def get_total_balance(self):
        self.total_balance = sum(account.get_balance() for account in self.accounts)
        return self.total_balance

    def find_account(self, account_holder_name):
        for account in self.accounts:
            if account.account_holder == account_holder_name:
                return account
        return None
    def reserve(self):
        self.get_total_balance()
        self.reserve=self.total_balance*.2
        return self.reserve
    def __str__(self):
        return f"{self.name}"
class Centralbank:
    def __init__(self):
        self.banks=[]
        self.total_reserve=0
    def add_bank(self,bank):
        if bank not in self.banks:
            self.banks.append(bank)
            print("%s certified member of central Bank"%bank)
        else:
            print(f"{bank} is already a member of Central Bank.")
        
    def calculate_total_reserve(self):
        for bank in self.banks:
            bank.reserve()
        self.total_reserve=sum(bank.reserve * 0.5 for bank in self.banks)
        return self.total_reserve

class Loan:
    def __init__(self,loan_amount,borrower_bank,lender_bank):
        self.borrower_bank=borrower_bank
        self.lender_bank=lender_bank
        self.loan_amount=loan_amount
    def borrow(self):
        if self.loan_amount>0:
            if self.borrower_bank in central_bank.banks and self.lender_bank in central_bank.banks:
                print(f"{self.lender_bank} is a certified member of Central Bank thus can borrow from reserved banks")
                self.lender_bank.reserve-=self.loan_amount
                self.borrower_bank.reserve+=self.loan_amount
                print(f"Loan of ${self.loan_amount} from {self.lender_bank} to {self.borrower_bank} successfully transferred.")
            else:
                print("Not allowed ,please register both banks to the Central Bank")
                sys.exit()
        else:
            print("Invalid amount")
                
               
                
            
        
        


bank1=Bank("bank1")
account1 = BankAccount("John Doe",2000.0)
account2 = BankAccount("Jane Smith", 1500.0)
bank2=Bank("bank2")
account3=BankAccount("Jordan Smith",54000.0)
account4 = BankAccount("Jane Patrick", 15000.0)

# Add accounts to the bank
bank1.add_account(account1)
bank1.add_account(account2)
bank2.add_account(account3)
bank2.add_account(account4)

# Deposit and withdraw from accounts
account1.deposit(500.0)
account2.withdraw(200.0)

# Check total balance in the bank
total_balance = bank1.get_total_balance()
print(f"Total balance in the bank: ${total_balance}")

#Checkout how much reserve in the central bank
central_bank=Centralbank()
central_bank.add_bank(bank1)
central_bank.add_bank(bank2)
print(f"Total reserve in central_bank: ${central_bank.calculate_total_reserve()}")

#loaning money
loan=Loan(10000,bank1,bank2)
loan.borrow()
# Check updated reserves after the loan
print(f"Bank1 reserve after loan: ${bank1.reserve}")
print(f"Bank2 reserve after loan: ${bank2.reserve}")

```
